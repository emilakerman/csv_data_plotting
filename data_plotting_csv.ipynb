```python
def read_file(file_name):
    data_list = []
    with open (file_name, "r", encoding = "utf-8-sig") as f:
        csv_reader = csv.reader(f, delimiter = ",")
    
        for row in csv_reader:
            data_list.append(row)
        return data_list
    

def calculate_mean(data):
    elements_in_list = 0
    for number in data:
        elements_in_list += number
    mean = elements_in_list / len(data)
    return mean
    
# Using multiple functions to deal with various calculations for better visibility    
def calculate_median(ladder, gdp, social, healthy, freedom, generosity, corruption):
    median_ladder = sorted(ladder)[68]
    median_logged = sorted(gdp)[68]
    median_social = sorted(social)[68]
    median_healthy = sorted(healthy)[68]
    median_freedom = sorted(freedom)[68]
    median_generosity = sorted(generosity)[68]
    median_corruption = sorted(corruption)[68]
    return median_ladder, median_logged, median_social, median_healthy, median_freedom, median_generosity, median_corruption

    
def calculate_min(ladder, gdp, social, healthy, freedom, generosity, corruption):
    min_ladder = sorted(ladder)[0]
    min_logged = sorted(gdp)[0]
    min_social = sorted(social)[0]
    min_healthy = sorted(healthy)[0]
    min_freedom = sorted(freedom)[0]
    min_generosity = sorted(generosity)[0]
    min_corruption = sorted(corruption)[0]
    return min_ladder, min_logged, min_social, min_healthy, min_freedom, min_generosity, min_corruption
    
    
def calculate_max(ladder, gdp, social, healthy, freedom, generosity, corruption):
    max_ladder = sorted(ladder)[-1]
    max_logged = sorted(gdp)[-1]
    max_social = sorted(social)[-1]
    max_healthy = sorted(healthy)[-1]
    max_freedom = sorted(freedom)[-1]
    max_generosity = sorted(generosity)[-1]
    max_corruption = sorted(corruption)[-1]
    return max_ladder, max_logged, max_social, max_healthy, max_freedom, max_generosity, max_corruption
    
     
def calculate_data(data):
    initial_rows = data[0]
    initial_rows.remove("Country name")
    
    ladder_score_data = [float(row[1]) for row in data[1:138]]
    logged_gdp_data = [float(row[2]) for row in data[1:138]]
    social_support_data = [float(row[3]) for row in data[1:138]]
    healthy_life_data = [float(row[4]) for row in data[1:138]]
    freedom_choice_data = [float(row[5]) for row in data[1:138]]
    generosity_data = [float(row[6]) for row in data[1:138]]
    corruption_perception_data = [float(row[7]) for row in data[1:138]]

    
    max_ladder_score, max_logged_gdp_data, max_social_support_data, max_healthy_life_data, max_freedom_choice_data, max_generosity_data, max_corruption_perception_data = calculate_max(ladder_score_data, logged_gdp_data, social_support_data, healthy_life_data, freedom_choice_data, generosity_data, corruption_perception_data)
    min_ladder_score, min_logged_gdp_data, min_social_support_data, min_healthy_life_data, min_freedom_choice_data, min_generosity_data, min_corruption_perception_data = calculate_min(ladder_score_data, logged_gdp_data, social_support_data, healthy_life_data, freedom_choice_data, generosity_data, corruption_perception_data)
    median_ladder_score, median_logged_gdp_data, median_social_support_data, median_healthy_life_data, median_freedom_choice_data, median_generosity_data, median_corruption_perception_data = calculate_median(ladder_score_data, logged_gdp_data, social_support_data, healthy_life_data, freedom_choice_data, generosity_data, corruption_perception_data)

    
    alla_medelvarden = [calculate_mean(ladder_score_data), calculate_mean(logged_gdp_data), calculate_mean(social_support_data), calculate_mean(healthy_life_data), calculate_mean(freedom_choice_data), calculate_mean(generosity_data), calculate_mean(corruption_perception_data)]
    alla_maxvarden = [max_ladder_score, max_logged_gdp_data, max_social_support_data, max_healthy_life_data, max_freedom_choice_data, max_generosity_data, max_corruption_perception_data]
    alla_minvarden = [min_ladder_score, min_logged_gdp_data, min_social_support_data, min_healthy_life_data, min_freedom_choice_data, min_generosity_data, min_corruption_perception_data]
    alla_medianer = [median_ladder_score, median_logged_gdp_data, median_social_support_data, median_healthy_life_data, median_freedom_choice_data, median_generosity_data, median_corruption_perception_data]
    
    
    return initial_rows, alla_medelvarden, alla_maxvarden, alla_minvarden, alla_medianer


def create_plotting(data, column, title):
    joined_list = [(row[0], float(row[column])) for row in data[1:9]] + [(row[0], float(row[column])) for row in data[-8:]]
    
    fig, ax = plt.subplots()

    countries, scores = zip(*joined_list)
    bar_colors = ("red", "blue", "orange", "green", "brown", "cyan", "purple", "pink", "grey", "salmon", "lightgreen", "indigo", "teal", "navy", "gold", "lime")

    ax.bar(countries, scores, color=bar_colors)

    plt.xlim(-1, len(countries))
    plt.ylim(0, max(scores)+0.4)

    ax.set_ylabel(title)
    ax.set_xlabel("Country")
    ax.set_title(f"{title} för år 2023")

    
    legend_patches = [mpatches.Patch(color = color, label = country) for country, color in zip(countries, bar_colors)]
    ax.legend(handles = legend_patches, loc = "upper left", prop = { "size": 8 })

    plt.xticks(rotation = 55)

    plt.show()

    
def create_divider():
    return "{0:30}{1}".format("+----------------------------------", "+------------+--------+--------+-------+")

    
    
WHR1Data = read_file("WHR2023.csv")

empty_string = ""

long_divider = "+----------------------------------+------------+--------+--------+-------+"

print(long_divider + "\n{0:74}{1}{2}".format("|År 2023", "|", f"\n{long_divider}"))
print("{0:35}{1:13}{2:9}{3:9}{4:8}{5}".format("|Namn", "|Medelvärde", "|Median", "|Min", "|Max", "|"))
print(long_divider)

initial_rows, alla_medelvarden, alla_maxvarden, alla_minvarden, alla_medianer = calculate_data(WHR1Data)

for i in range(len(initial_rows)):
    name = initial_rows[i]
    medel = round(alla_medelvarden[i], 3)
    max_varde = alla_maxvarden[i]
    min_varde = alla_minvarden[i]
    median = alla_medianer[i]
    print("{0:35}{1:13}{2:9}{3:9}{4:8}".format(f"|{name}", f"|{medel}", f"|{median}", f"|{min_varde}", f"|{max_varde}"), end=f"|\n{create_divider()}\n")
    
for i in range(7):
    create_plotting(WHR1Data, i+1, WHR1Data[0][i])    

```

```python
def read_file(file_name):
    data_list = []
    with open (file_name, "r", encoding = "utf-8-sig") as f:
        csv_reader = csv.reader(f, delimiter = ",")
    
        for row in csv_reader:
            data_list.append(row)
        return data_list
    
    
def extract_relevant_data(data, region):
    relevant_data_2021 = [(index, row[0], row[1], float(row[10])) for index, row in enumerate(data[1:147]) if row[1] == str(region)]
    relevant_data_2020 = [(index, row[0], row[1], float(row[11])) for index, row in enumerate(data[1:147]) if row[1] == str(region)]
    
    return relevant_data_2021, relevant_data_2020
    

def calculate_mean(happiness_data_2021, happiness_data_2020):
    elements_in_list = 0
    for index, country, region, data in happiness_data_2021:
        elements_in_list += data
    for index, country, region, data in happiness_data_2020:
        elements_in_list += data
    mean = elements_in_list / (len(happiness_data_2020) + len(happiness_data_2021))
    return mean


def create_plotting(data_2021, data_2020, slicing):
    if slicing == "top_5":
        top_or_bottom = "högsta"
        country_list_2021 = data_2021[0:5]
        country_list_2020 = data_2020[0:5]
    elif slicing == "bottom_5":
        top_or_bottom = "lägsta"
        country_list_2021 = data_2021[-5:]
        country_list_2020 = data_2020[-5:]
    
    fig, ax = plt.subplots()

    index, countries, region, happiness_data = zip(*country_list_2021)
    index, countries_2020, region, happiness_data_2020 = zip(*country_list_2020)
    bar_colors_2021 = ("red", "blue", "orange", "green", "brown")
    bar_colors_2020 = ("lightcoral", "lightblue", "gold", "lightgreen", "teal")
    circle_colors_2021 = ("lime", "magenta", "pink", "silver", "salmon")
    circle_colors_2020 = ("gray", "sienna", "green", "purple", "yellow")
    
    # Different bar color lists so they dont have the same colors
    ax.bar(countries_2020, happiness_data_2020, color = bar_colors_2020, width = 0.2, align = "center")
    ax.bar(countries, happiness_data, color = bar_colors_2021, width = 0.1, align = "center")

    # Xlim decimals like this to make sure everything is visible and looks good
    plt.xlim(-0.3,4.3)
    plt.ylim(0,28)

    ax.set_ylabel("happiness2021, happiness2020 och rank")
    ax.set_xlabel("Country")
    ax.set_title(f"5 länder med {top_or_bottom} happiness2021 och happiness2020 samt rank för Asia")

    legend_patches = [
        mpatches.Patch(color = bar_colors_2021[i], label = f"{countries[i]} 2021")
        for i in range(len(countries))
    ]
    legend_patches += [
        mpatches.Patch(color = bar_colors_2020[i], label = f"{countries_2020[i]} 2020")
        for i in range(len(countries_2020))
    ]
    legend_patches += [
        mpatches.Patch(color = circle_colors_2021[i], label = f"{countries[i]} 2021")
        for i in range(len(countries))
    ]
    legend_patches += [
        mpatches.Patch(color = circle_colors_2020[i], label = f"{countries_2020[i]} 2020")
        for i in range(len(countries_2020))
    ]
    ax.legend(handles = legend_patches, loc = "upper right", prop = {"size": 6})
    
    # Creation and plotting of circles using x_axis_number to assist with location 
    x_axis_number = 0
    for (index_2021, country_2021, region_2021, happiness_data_2021), (index_2020, country_2020, region_2020, happiness_data_2020) in zip(country_list_2021, country_list_2020):
        _ = plt.Circle((x_axis_number, index_2021+1), 0.7, color = circle_colors_2021[x_axis_number])
        __ = plt.Circle((x_axis_number, index_2020+1), 0.7, color = circle_colors_2020[x_axis_number])
        _.height = 3
        _.width = 0.4
        __.height = 2
        __.width = 0.25
        ax.add_patch(_)
        ax.add_patch(__)
        x_axis_number = x_axis_number + 1

    plt.xticks(rotation = 30)
    plt.grid()
    plt.show()
        

WHR2Data = read_file("WHR2020-2021.csv")
    
region_choice = str(input("Ange region: "))
print("\n---------------------------------------------")
happiness_data_2021, happiness_data_2020 = extract_relevant_data(WHR2Data, region_choice)
mean = calculate_mean(happiness_data_2021, happiness_data_2020)
print(f"Medelvärde {round(mean,2)} över happiness index för länder i {region_choice} år 2021 och 2020\n")
print(f"Länder i {region_choice}:\n")

divider = "+--------------------------+---------------+----------------+-----+"
print("{0:27}{1:16}{2:17}{3:6}{4:5}".format(f"|Land", f"|Score för 2021", f"|Score för 2020", f"|Rank", f"|"))
print(divider)
for (index_2021, country_2021, region_2021, data_2021), (index_2020, country_2020, region_2020, data_2020) in zip(happiness_data_2021, happiness_data_2020):
    print("{0:27}{1:16}{2:17}{3:6}{4:10}".format(f"|{country_2021}", f"|{data_2021}", f"|{data_2020}", f"|{index_2021+1}", f"|"))
    print(divider)
    
create_plotting(happiness_data_2021, happiness_data_2020, "top_5")
create_plotting(happiness_data_2021, happiness_data_2020, "bottom_5")

```

```python
def read_file(file_name):
    data_list = []
    with open (file_name, "r", encoding = "UTF-8") as f:
        csv_reader = csv.reader(f, delimiter = ",")
    
        for row in csv_reader:
            data_list.append(row)
        return data_list
    
    
def extract_country_data(data, country):
    life_ladder_data = [(int(row[1]), float(row[2])) for row in data if row[0] == str(country)]
    return life_ladder_data


def calculate_mean(data):
    elements_in_list = 0
    for year, number in data:
        elements_in_list += number
    mean = elements_in_list / len(data)
    return mean


def draw_line_chart(data, country, years):
    fig, ax = plt.subplots()
    data_values = []
    for year, number in data:
        data_values.append(number)
        
    ax.set_ylabel("Life ladder")
    ax.set_xlabel("year")
    ax.set_title(f"Life ladder for {country}")
    
    legend_patches = [
        mpatches.Patch(color = "blue", label = f"{country}")
    ]
    ax.legend(handles = legend_patches, loc = "upper right", prop = {"size": 8})
    
    plt.xlim(years[0],years[-1])
    
    x = np.array(years)
    y = np.array(data_values) 

    
    plt.plot(x, y) 
    plt.grid()
    plt.show()


def draw_bar_chart(data, changes, country, years):
    fig, ax = plt.subplots()
    
    years = years[ : -1]
    
    bar_colors = ("red", "blue", "orange", "green", "brown", "cyan", "purple", "pink", "grey", "salmon", "lightgreen", "indigo", "teal", "navy", "gold", "lime")

    ax.bar(years, changes, color = bar_colors, align = "center")

    plt.xlim(years[0], years[-1])
    plt.ylim(-4, 4)

    ax.set_ylabel("Förändringsfaktor")
    ax.set_xlabel("year")
    ax.set_title(f"Förändringsfaktor för {country} över åren i procentform")

    legend_patches = [mpatches.Patch(color = color, label = year) for year, color in zip(years, bar_colors)]
    ax.legend(handles = legend_patches, loc = "upper left", prop = {"size": 8})
    
    plt.xticks(rotation = 55)
    plt.grid()
    plt.show()


WHR3Data = read_file("WHR2005-2022.csv")

country_choice = str(input("Ange landet som ska presenteras: "))
country_life_ladder_data = extract_country_data(WHR3Data, country_choice)
mean = calculate_mean(country_life_ladder_data)
print(f"Medelvärde av Life Ladder är {round(mean,1)} för {country_choice}")

# Final fix to my issue with variable data lengths of countries
line_years = []
for year, number in country_life_ladder_data:
    line_years.append(year)
draw_line_chart(country_life_ladder_data, country_choice, line_years)


data_list = country_life_ladder_data
# Changing order/rotating(?) the elements one step to the left to compare values in the for loop
data_list = data_list[1:] + [data_list[0]]      
    
divider = "+-------+----------------+------------------"    
    
print("{0:8}{1:15}{2:15}".format(f"|År", f"|score för {country_choice}", f"|Förändringsfaktor"))
print(divider)
percentage_changes = []
for (year, data), (year_2, data_2) in zip(country_life_ladder_data, data_list):
    # Updated calculation with correct FF
    percentage_change = (data_2 / data)
    percentage_change = 1 - percentage_change
    percentage_change = (percentage_change * 100) * -1
    percentage_changes.append(round(percentage_change,2))
    print("{0:8}{1:17}{2:16}".format(f"|{year}", f"|{data}", f"|{round(percentage_change, 2)}"))
    print(divider)

# Removing last element in list due to example graph not showing year 2022
draw_bar_chart(country_life_ladder_data, percentage_changes[:-1], country_choice, line_years)
```
